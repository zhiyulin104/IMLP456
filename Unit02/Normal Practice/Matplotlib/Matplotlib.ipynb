# 繪製基本圖表 

* 匯入`Matplotlib`套件的`pyplot`模組，別名`plt`

```python
import matplotlib.pyplot as plt
import numpy as np
```

* 繪製簡單的折線圖

```python
data = [20.1, 22.2, 25.5, 27.7, 30.4, 33.1, 34.9, 36.0, 35.5, 32.0, 28.3, 24.6]
labels = range(1, 13)
plt.plot(labels, data)
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Average Monthly Temperature')
plt.show()
```

* 繪製不同線條樣式和色彩的折線圖 
    - 請修改折線圖的線條外觀，改為藍色虛線，和加上圓形標記

```python
plt.plot(labels, data, marker='^', linestyle='-.', color='orange')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Average Monthly Temperature with Markers')
plt.grid(True)
plt.show()
```

* ## 額外

```python
data = [20.1, 22.2, 25.5, 27.7, 30.4, 33.1, 34.9, 36.0, 35.5, 32.0, 28.3, 24.6]
data2 = [18.3, 20.1, 23.4, 26.1, 28.9, 31.0, 33.5, 34.2, 33.8, 30.5, 27.0, 22.8]

def plot_temperature(data, color, label) -> list:
    labels = range(1, 13)
    i = 0
    for d in data: 
        plt.plot(labels, d, marker='o', linestyle='--', color = color[i], label = label[i])
        i += 1
    xmin, xmax, ymin, ymax = 0.5, 12.5, 15, 40
    plt.xlabel('Month')
    plt.ylabel('Temperature (°C)')
    plt.title('Average Monthly Temperature')
    plt.axis([xmin, xmax, ymin, ymax])
    plt.grid(True)
    plt.legend(loc = 4)
    plt.show()
    
plot_temperature([data, data2], ['orange', 'blue'], ['City A', 'City B'])
```

* 繪製每日攝氏溫度的折線圖

```python
days = range(1, 8)
celcius = [22, 24, 19, 23, 25, 21, 20]
plt.plot(days, celcius)
plt.show()
```

* 使用多個資料集繪製多條折線 
    - 使用`NumPy`陣列作，這是使用`np.sin()`和`np.cos()`方法建立的2個資料集，可以在同 一張圖表繪出2條折線

```python
x = np.linspace(min(data), max(data), 100)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, x, cosinus)
plt.title('Sine Function')
plt.xlabel('X-axis')
plt.ylabel('sin(X)')
plt.show()
```

# 繪製基本圖表

* 顯示圖表的格線 
    - `Matplotlib`可以使用`grid()`方法切換顯示圖表的水平和垂直格線（參數值True）

```python
days = range(1, 8)
celcius = [22, 24, 19, 23, 25, 27, 26]
plt.plot(days, celcius, "r-o")
plt.grid(True)
plt.show()
```

* 顯示x和y軸的說明標籤 
    - 在`Matplotlib`的`x`軸使用`xlabel()`方法指定標籤“日數”和`ylabel()`方法指定y軸標籤 “攝氏溫度”。

```python
days = range(1, 8)
celsius = [22, 24, 19, 23, 25, 27, 26]
plt.plot(days, celsius, "g--s")
plt.xlabel("Days")
plt.ylabel("Celsius")
plt.title("Weekly Celsius Trend")
plt.grid(True)
plt.show()
```

* 顯示圖表的標題文字 
    - 請使用`title()`方法指定圖表標題文 字“Sin 和 Cos 波形”

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o",
         x, cosinus, "g--")
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("Sin 和 Cos 波形")
plt.show()
```

* 顯示圖例 
    - 如果在同一張圖表繪出多條線， `Matplotlib`可以顯示圖例（`Legend`）來標示每一條線是哪 一個資料集。例如：在圖表顯示 圖例標示2條線分別是`sin(x)`和 `cos(x)`三角函數，如右所示

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o", label = "sin(x)")
plt.plot(x, cosinus, "g--", label = "cos(x)")
plt.legend(loc = 1)
plt.show()
```

* 顯示軸的範圍 
    - 請使用`axis()`方法顯示`Matplotlib` 自動計算出的軸範圍，如右圖所示： 

* 指定軸的自訂範圍 
    - 如果覺得自動計算出的軸範圍並不符合預期，可以使用`axis()`方法自行指定x和y軸的範圍，如下圖所示:

```python
days = range(1, 8)
celsius = [22, 24, 19, 23, 25, 27, 26]
plt.plot(days, celsius, "g--s")
print(plt.axis())  # 顯示目前的軸範圍
plt.show()
```

```python
days = range(1, 8)
celsius = [22, 24, 19, 23, 25, 27, 26]
plt.plot(days, celsius, "g--s")
print(plt.axis())  # 顯示目前的軸範圍
xmin, xmax, ymin, ymax = 0, 8, 15, 30
plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

* 指定多個資料集的軸範圍
    - 如果是多個資料集的圖表，一樣可 以使用`axis()`方法指定自訂的x和y 軸範圍，如右所示：

```python
days = range(1, 8)
celsius_min = [22, 21, 19, 20, 22, 23, 21]
celsius_max = [25, 26, 24, 27, 28, 29, 26]
plt.plot(days, celsius_min, "r-o",
         days, celsius_max, "g--o")
plt.xlabel("Days")
plt.ylabel("Celsius")
plt.title("Weekly Celsius Range")
plt.axis([0.5, 7.5, 15, 35])
plt.legend(["Min Temp", "Max Temp"], loc=1)
plt.show()
```

## 1.散佈圖

* 繪製`Sin()`三角函數的散佈圖 
    - `Matplotlib`是呼叫`scatter()`方法繪製散佈圖，散佈圖就是點的集合，在各點之間 沒有連線，例如：將`y=sin(x)`建立成散佈圖，如下所示：

```python
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

## 2.長條圖 

* 「長條圖」（`Bar Plots`） 
    - 使用長條型色彩區塊的高和長度來顯 示分類資料，分為水平或垂直方向的長條。 
    - 在這一節準備使用`TIOBE`常用程式語 言的使用率來繪製長條圖，首先建立4個清單和`NumPy`陣列。 
    - `Matplotlib`是呼叫`bar()`方法繪製長條 圖，第1個參數是x軸的`index`索引， 第2個y軸資料是`ratings`使用率。

```python
import warnings
warnings.filterwarnings('ignore')

labels = ["Python","C++","Java","JS","C","C#"]
index = np.arange(len(labels))
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
change = [1.12,0.3,-1.69,0.29,3.41,-0.45]
```

```python
plt.bar(index,ratings)
plt.xticks(index,labels)
plt.ylabel("Change")
plt.title("Programming rating")
plt.show()
```

* 水平長條圖 
    - `Matplotlib`只需改用`barh()`方法，就可以繪製成水平長條圖，如下所示：

```python
plt.barh(index,ratings)
plt.yticks(index,labels)
plt.xlabel("ratings")
plt.title("Programming rating")
plt.show()
```

## 直方圖 

* 顯示直方圖的區間和出現次數 
    - 直方圖（`Histograms`）是用來顯示 數值資料的分佈，一種次數分配表， 可以使用長方形面積顯示變數出現的 頻率，寬度是分割區間 。 

    
    - 請使用整數清單（共21個元素）顯 示直方圖的區間和出現次數（即每一 個區間的次數分配表），這是呼叫 `hist()`方法繪製直方圖，第 1個參數 是資料清單或NumPy陣列，第 2 個 參數是分割成幾個區間，以此例是 5 個，方法回傳的 n是各區間的出現次 數，bins是分割 5個區間的值，如右 所示：

```python
x = [21,42,23,4,5,26,77,88,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5
n, bins, patches = plt.hist(x,num_bins)
print(n)
print(bins)
print(patches)
plt.show()
```

## 派圖 

* 派圖（`Pie Plots`）也稱為圓餅圖（`Circle Plots`） 
    - 使用完整圓形來表示統計資料的圖表，如同切圓形蛋榚，以不同切片大小來標示資料 的比例。 – 也是使用TIOBE常用程式語言的使用率來繪製派圖，首先建立2個清單，`labels`清單 是語言標籤；`ratings`清單是對應各種語言的使用率，如下所示

```python
labels = ["Python","C++","Java","JS","C","C#"]
ratings = [5.16,5.73,14.99,3.17,11.86,4.45]
plt.pie(ratings, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Programming Language Popularity")
plt.show()
```
