## SERIES物件 
* 因為`Pandas`套件關於資料處理和分析的重點是`DataFrame`物件，所以本 章只準備簡單說明`Series`物件的使用 
* 在`Python`程式首先需要匯入`Pandas`套件的別名`pd`，如下所示：

```python
import pandas as pd
```

* 建立`Series`物件 
    - `Series`物件可以使用`Python`清單來建立，使用的是`Series()`方法，參數是`Python` 清單，如下所示：

```python
s = pd.Series([12, 29, 72, 4])
print(s)
```

* 建立自訂索引的`Series`物件 
    - `Series`物件如同是2個陣列，一是索引標籤；一是資料，所以可以使用2個`Python` 清單建立`Series`物件，在`Series()`方法的第1個參數是資料清單，第2個是使用 `index`參數指定的索引清單 
    - 然後使用`index`屬性顯示索引；`values`屬性可以顯示資料

```python
fruits = ["蘋果", "橘子", "梨子"]
quantities = [15, 33, 45]
s = pd.Series(quantities, index=fruits)
print(s)
```

```python
print(s.index)
print(s.values)
```

* 使用索引取出資料和執行運算 
    - 在建立`Series`物件後，可以使用索引值來取出資料 
    - 如同`NumPy`陣列，一樣可以使用索引清單一次就取出多筆資料 
    - `Series`物件也可以作為運算元來執行四則運算，如下所示

```python
print(s["橘子"])
```

```python
print(s[["橘子", "梨子"]])
```

```python
print((s + 2) * 3)
```

# 建立DATAFRAME物件

* 使用`Python`字典建立`DataFrame`物件 
    - `DataFrame`物件是表格擁有列和欄標籤，首先建立3個元素的`fruits`字典，鍵是字串； 值是清單，然後呼叫`pd.DataFrame()`方法建立`DataFrame`物件，如下所示：

```python
fruits = {"蘋果": [4, 3, 1, 0],
          "香蕉": [0, 4, 6, 2],
          "橘子": [1, 5, 2, 4]}

fruit_df = pd.DataFrame(fruits)
fruit_df
```

* 使用自訂索引建立`DataFrame`物件 
    - 如果沒有指明索引標籤，`Pandas`預設替`DataFrame`物件產生數值索引標籤，當然 可以使用`cites`清單來建立自訂索引標籤，共有4個元素，對應4筆資料。
    - 在建立`DataFrame`物件時，可以使用`columns`參數**重新指定欄位順序**，將原來蘋果、 香蕉、橘子順序改為香蕉、橘子、蘋果。

```python
cites = ["台北", "新北", "台中", "高雄"]
df = pd.DataFrame(fruits, index=cites)
df
```

```python
df = pd.DataFrame(fruits,
                  columns=["橘子", "蘋果", "香蕉"],
                  index=["台北", "新北", "台中", "高雄"])
df 
```

* 更改索引標籤和欄位標籤 
    - 在建立`DataFrame`物件`df`後，可以使用`columns`屬性重新指定欄位標籤，`index`屬 性是更改索引標籤，如下所示：

```python
df.columns = ["banana", "orange", "apple"]
cites[2] = "桃園"
df.index = cites
df
```

* 轉置`DataFrame`物件 
    - 可以使用`.T`屬性轉置`DataFrame`物件，即欄變列、列成欄

```python
df.T
```

## 選取資料 
* 使用索引標籤選取特定範圍的記錄 
    - 對於`DataFrame`物件每一列的記錄來說，可以使用從0開始的索引標籤，或使用自 訂索引標籤來選取特定範圍的記錄，首先是預設數字索引的範圍，可以選取第1~2 筆記錄，但不含索引值2的第3筆 
    - 如果使用自訂索引標籤，例如：索引"C"到"E"範圍，就會包含最後一筆，即"E"

```python
df[0:2]
```

```python
df["C":"E"]
```

* 使用loc索引器選取資料 
    - 在DataFrame二維表格定位資料時，loc索引器可以使用索引和欄位標籤來取出表格的 部分資料，其語法如下所示： 
    >df.loc[索引標籤] <br>
    df.loc[[索引標籤1, 索引標籤2,…]]<br>
    df.loc[索引標籤, 欄位標籤] <br>
    df.loc[[索引標籤1, 索引標籤2,…], [欄位標籤1, 欄位標籤2…]] <br>
    df.loc[索引標籤1:索引標籤2, 欄位標籤] <br>
    df.loc[索引標籤1:索引標籤2, [欄位標籤1, 欄位標籤2…]] <br>
    df.loc[索引標籤1:索引標籤2, 欄位標籤1:欄位標籤2] <br>

* 使用iloc索引器選取資料 
    - DataFrame物件的iloc索引器是使用從0開始的索引位置值來選取資料，其語法如下所示： 
    df.iloc[列索引位置, 欄索引位置]

* 使用loc索引器選取資料

## 過濾資料 
* 使用布林條件和`isin()`方法來過濾資料 
    - 在`DataFrame`物件的欄位標籤可以使用布林條件，只選擇條件成立的記錄資料。 
* 使用多個條件來過濾資料 
    - 在布林索引可以使用多個條件，例如：`sales`大於15，且小於25，這是使用「&」 的And邏輯運算子，如下所示：

```python
df[df.sales > 15]
```

```python
df[(df.sales > 15) & (df.sales < 25)]
```

## 排序資料 
* 指定欄位標籤來排序記錄資料 
    - `DataFrame`物件可以呼叫`sort_values()`方法，指定特定欄位標籤來排序記錄資料。

```python
df.sort_values("sales", ascending=False)
```

* 排序索引或欄位標籤 
    - `DataFrame`物件使用`set_index()`方法可以指定索引標籤是哪一個欄位標籤後，即可 呼叫`sort_index()`方法排序目前的索引標籤，看到索引標籤從大至小排序，因為 axis屬性值是0。

```python
df.set_index("sales", inplace=True)
df.sort_index(axis=0, ascending=True)
```

# 群組、樞紐分析與統計函數 

* 「群組」（Grouping）是先將資料依條件分類成群組後，再套用相關方 法在各群組來取得統計資料。欲使用群組來計算總計，首先使用字典建立 測試的`DataFrame`物件，如下所示

![](pic5.png)

```python
data = {"名稱":["客戶A","客戶B","客戶A","客戶B","客戶A","客戶B","客戶A","客戶A",], 
        "編號":["訂單1","訂單1","訂單2","訂單3","訂單2","訂單2","訂單1","訂單3"], 
        "數量":[4,4,1,2,3,4,2,1], 
        "售價":[495,496,360,451,221,321,466,260]}

df = pd.DataFrame(data)
df
```

## 群組
* 群組 
    - 名稱和編號欄位都有重複資料，我們可以分別使用這些欄位來群組記錄後，呼叫 `sum()`方法計算欄位值的總和。

```python
df.groupby("名稱").sum()
```

## 樞紐分析表

* 樞紐分析表 
    - `DataFrame`物件是呼叫`pivot_table()`方法產生樞紐分析表，資料為左下表格新增 `score`欄位。

```python
products = {"channel": ["網路","網路", "電視", "電視", "郵購", "郵購"],
            "company": ["EHS", "Momo", "EHS", "Viva", "Momo", "EHS"],
            "sales": [11.22, 23.50, 12.99, 15.95, 25.75, 11.55]}
oridinal = ["A", "B", "C", "D", "E", "F"]
df2 = pd.DataFrame(products, index=oridinal)
df2
```

```python
score = [4, 5, 6, 7, 5, 8]
df2["score"] = score
df2
```

* 使用`pivot_table()`方法以欄位值為標籤來重塑`DataFrame`物件的形狀， 也就是建立樞紐分析表

```python
pivot_tb = df2.pivot_table(index="channel",
                          columns="company",
                          values="sales")

pivot_tb
```

# PANDAS 資料視覺化

* 建立`DataFrame`或`Series`物件後，就可以使用`plot()`方法繪製長條圖

```python
df = pd.read_csv("Kobe_stats.csv")
data = pd.DataFrame()
data["Season"] = pd.to_datetime(df["Season"])
data["PTS"] = df["PTS"]
data["AST"] = df["AST"]
data["REB"] = df["REB"]
data = data.set_index("Season")
data.plot(title="Kobe Bryant Season Stats", figsize=(10,6), kind="line")
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(100, 4), index=pd.date_range('1/1/2024', periods=100), columns=list('ABCD'))

df = df.cumsum()
df.plot(title="Random Data Cumulative Sum", figsize=(10,6), kind="line")
```
