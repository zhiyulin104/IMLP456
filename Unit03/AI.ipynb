# ARTIFICIAL INTELLIGENCE 

## Can Machines Think?
**一套評估機器能否思考的方法: 圖靈測試 (Turing Test)** 

* 弱人工智慧 (Weak AI) 指人工模擬智慧，主張機器只能模擬人類具有思維的行為表現，而不是 真的懂得思考(無創造力) 

* 強人工智慧 (Strong AI) 跟人一樣有很廣泛的智慧(思考能力) ， 這是AI發展目標，尚未實現
